To start with, let's load the "breast cancer" data set from scikit-learn:

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
y = np.matrix(data.target).T
X = np.matrix(data.data)
M = X.shape[0]
N = X.shape[1]

# Normalize each input feature
def normalize(X):
    M = X.shape[0]
    XX = X - np.tile(np.mean(X,0),[M,1])
    XX = np.divide(XX, np.tile(np.std(XX,0),[M,1]))
    return XX

XX = normalize(X)
```

Next, let's represent a fully-connected neural network by two arrays W and b containing the weights and biases for each layer.

```python
# units is the number of units each layer, e.g. [N,h,1] is input layer N unit, second layer h ...
def initializer(no_of_layers, units):
    for i in range(0,no_of_layers-1):
        W.append(np.random.normal(0,0.1,[units[i],units[i+1]]))
        b.append(np.random.normal(0,0.1,[units[i+1],1]))

#sigmoid        
def actSig(z):
    return 1/(1+np.exp(-z))

#activation derivative
#sigmoid
def actSigder(z):
    az = actSig(z)
    prod = np.multiply(az,1-az)
    return prod

#ReLU
def actReLU(z):
    for i in range(len(z)):
        z[i] = z[i] * (z[i] >= 0)
    return z

#ReLU derivative
def actReLUder(z):
    az = actReLU(z)
    print(az)
    for i in range(len(z)):
        z[i] = 1 * (az[i] > 0) 
    return z

def ff(x,W,b):
    L = len(W)-1
    a = x
    for l in range(1,L+1):
        z = W[l].T*a+b[l]
        a = act(z)
    return a

def loss(y,yhat):
    return -(np.multiply((1-y), np.log(1-yhat)) + np.multiply(y , np.log(yhat)))
```

## Question 2

Beginning with the training code we wrote together in class, write Python code to execute backpropagation with mini-batch size 1 on the training set, and plot the training loss and validation loss as a function of training iteration. Show the plot in this sheet.

```python
#batch_size
batch_size = 2

#initializing number of layers and units
h1 = 10
W = [[]]
b = [[]]

initializer(3,[N,h1,1])
L = len(W)-1

alpha = 0.01

#max_iter
max_iter = 10

def ff(L,X_t,split_index,batch_size,y_t=None,type="trainer",hidden_layer_act="actSig"):
    loss_this_iter = 0
    order = np.random.permutation(split_index)

    for i in range(0,int(split_index/batch_size)):
        # Grab the pattern order[i]

        x_this = X_t[order[i:i+batch_size],:].T
        if y_t.all:
            y_this = y_t[order[i:i+batch_size],:]

        # Feed forward step

        a = [x_this]
        z = [[]]
        delta = [[]]
        dW = [[]]
        db = [[]]
        for l in range(1,L+1):
            z.append(W[l].T * a[l-1] + b[l])
            if hidden_layer_act == "actSig":
                a.append(actSig(z[l]))
            elif hidden_layer_act == "actReLU" and l < L:
                a.append(actReLU(z[l]))
            else:
                a.append(actSig(z[l]))
                
            # Just to give arrays the right shape for the backprop step
            delta.append([]); dW.append([]); db.append([])

        loss_this_pattern = np.sum(loss(y_this, a[L].T),axis=0)
        loss_this_iter = loss_this_iter + loss_this_pattern
        
        if type == "trainer":
            if hidden_layer_act == "actSig":
                backprop(L,a,z,delta,dW,db,y_this)
            elif hidden_layer_act == "actReLU":
                backprop(L,a,z,delta,dW,db,y_this,hidden_layer_act="actReLU")
            
    return loss_this_iter[0,0]
        
        
# Backprop step
def backprop(L,a,z,delta,dW,db,y_this,hidden_layer_act="actSig"):        
    delta[L] = np.sum(a[L].transpose() - y_this, axis=0)
    for l in range(L,0,-1):
        db[l] = delta[l].copy()
        dW[l] = np.sum(a[l-1],axis=1) * delta[l].transpose()
        if l > 1:
            delta[l-1] = np.sum(np.multiply(actSigder(z[l-1]) , W[l] * delta[l]),axis=1)
            print(delta[l-1].shape)

    for l in range(1,L+1):            
        W[l] = W[l] - alpha * dW[l]
        b[l] = b[l] - alpha * db[l]

#this function is for testing and validation     
def trainer(W,b,L,X_t,y_t,split_index,batch_size,X_v,y_v,max_iter=1000,hidden_layer_act="actSig"):
    print("Please Wait...")
    #loss in training set every iteration
    loss_every_iter_train = []
    
    #loss in validation set every iteration
    loss_every_iter_val = []
    
    for iter in range(0, max_iter):
        
        #for training set
        if hidden_layer_act == "actSig":
            loss_this_iter = ff(L,X_t,split_index,batch_size,y_t,"trainer")
        elif hidden_layer_act == "actReLU":
            loss_this_iter = ff(L,X_t,split_index,batch_size,y_t,"trainer",hidden_layer_act="actReLU")
        loss_every_iter_train.append(loss_this_iter)
        
        #for validation set
        if hidden_layer_act == "actSig":
            loss_this_iter = ff(L,X_v,len(X)-split_index,batch_size,y_v)
        elif hidden_layer_act == "actReLU":
            loss_this_iter = ff(L,X_v,len(X)-split_index,1,y_v,hidden_layer_act="actReLU")
        loss_every_iter_val.append(loss_this_iter)
    
    #for training set
    plt.title("Iterations Vs Loss\n Test: Blue, Validation: Orange")
    plt.plot(range(0,len(loss_every_iter_train)),loss_every_iter_train)
    plt.plot(range(0,len(loss_every_iter_val)),loss_every_iter_val)
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    print("Loss in Training set:",loss_every_iter_train[len(loss_every_iter_train)-1])
    print("Loss in validation set:",loss_every_iter_val[len(loss_every_iter_val)-1])
    plt.show()

trainer(W,b,L,X_train,y_train,split_index,batch_size,X_validate,y_validate,max_iter)
```
